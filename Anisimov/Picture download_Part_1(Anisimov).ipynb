```python
# /extract crop,resize and save final version of pictures
def save_crop(path_final,path_collection):
    b=0
    for i in path_collection:
        try:
            im=Image.open(i)

            im=im.convert('RGB')
            w,h=im.size
            im=im.crop((0,0,w,h*0.94))
            w,h=im.size
            J=0.97
            c_w=w/2
            c_h=h/2
            if h>w:
                im=im.crop((c_w-c_w*J,c_h-c_w*J,c_w+c_w*J,c_h+c_w*J))
            else:
                im=im.crop((c_w-c_h*J,c_h-c_h*J,c_w+c_h*J,c_h+c_h*J))
            im=im.resize((512,512))
            b+=1
            im.save(str(path_final)+str(b)+'.jpg')
        except:
            pass
```
